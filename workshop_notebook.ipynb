```python
genre_data = genre_data.iloc[0:23]
```

```python
# 
genre_data['title'] = genre_data['title'].apply(lambda x: x.lower())
genre_data['year'] = genre_data['year'].apply(lambda x: int(x))
```

```python
genre_data[genre_data['title']=='fight club']
```

```python
imdb_min = genre_data['imdb rating'].min()
imdb_max = genre_data['imdb rating'].max()
genre_data['imdb rating'] = genre_data['imdb rating'].apply(lambda x: (x - imdb_min)/(imdb_max - imdb_min))

rt_min = genre_data['rotten tomatoes rating'].min()
rt_max = genre_data['rotten tomatoes rating'].max()
genre_data['rotten tomatoes rating'] = genre_data['rotten tomatoes rating'].apply(lambda x: (x - rt_min)/(rt_max - rt_min))
```

```python
genre_data
```

```python
# alle Spalten der Tabelle, die nicht der Filmtitel sind
genres = [i for i in genre_data.columns if i not in ['title', 'year', 'imdb rating', 'rotten tomatoes rating']]
ratings = ['imdb rating', 'rotten tomatoes rating']
year = ['year']
```

```python
genre_data['total_rating'] = genre_data['imdb rating'] + genre_data['rotten tomatoes rating']
genre_data['mean_rating'] = 0.5 * (genre_data['imdb rating'] + genre_data['rotten tomatoes rating'])
genre_data[['title', 'mean_rating', 'total_rating']].sort_values(by='mean_rating', ascending=False)
```

```python
by_genre = genre_data[genres]
by_rating = genre_data[ratings]
by_everything = genre_data[[i for i in genre_data.columns if 'title' not in i]]
titles = genre_data.title
```

```python
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
```

```python
n_clusters = 3
cluster = KMeans(n_clusters=n_clusters)
genre_clustered = cluster.fit_predict(by_genre)
genre_distance = cluster.transform(by_genre)
rating_clustered = cluster.fit_predict(by_rating)
rating_distance = cluster.transform(by_rating)
everything_clustered = cluster.fit_predict(by_everything)
everything_distance = cluster.transform(by_everything)
genre_data['by_rating'] = rating_clustered
genre_data['by_genre'] = genre_clustered
genre_data['by_everything'] = everything_clustered

print('genres:')
for i in range(n_clusters):
    print(i)
    indices = np.where(genre_clustered==i)[0]
    print(titles[indices].values)

print()
print('ratings:')
for i in range(n_clusters):
    print(i)
    indices = np.where(genre_clustered==i)[0]
    print(titles[indices].values)

print()
print('everything:')
for i in range(n_clusters):
    print(i)
    indices = np.where(everything_clustered==i)[0]
    print(titles[indices].values)
```
