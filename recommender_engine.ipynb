```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# unsere bearbeiteten Datensätze
user_distances = pd.read_csv('user_distances.csv')
movie_genres = pd.read_csv('filme_mit_cluster.csv')
user_data = pd.read_csv('workshop_users.csv')
```

```python
# fügen wir die cluster und die user ratings informationen zusammen
user_data = user_data.join(movie_genres.set_index('title')['cluster'], on='title', how='left')
```

```python
user_id = 'user10'
# finde den User, der user10 am ähnlichsten ist
# da der User sich selbst am ähnlichsten ist, nehmen wir den zweiten (bzw. da Python 0-basiert ist, den 1) index
closest_user = user_distances[user_id].sort_values().index[1]
print('Der {} ähnlichste User ist {}.'.format(user_id, closest_user))
user_data[['title', 'cluster', user_id, 'user{}'.format(closest_user)]]
```
